```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('company_sales_data - company_sales_data.csv')
features = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
sales
```

```python
month = sales['month_number']
profit = sales['total_profit']
```

```python
plt.plot(month,profit)
plt.xlabel('months')
plt.ylabel('units')
```

```python
plt.plot(month, sales[features])
plt.xlabel('Months')
plt.ylabel('Units sold')
plt.legend(features)
```

```python
cream_sum = sales.facecream.sum()
wash_sum = sales.facewash.sum()
tooth_sum = sales.toothpaste.sum()
soap_sum = sales.bathingsoap.sum()
shampoo_sum = sales.shampoo.sum()
lotion_sum = sales.moisturizer.sum()
units_sum = sales.total_units.sum()
```

```python
cream_size = cream_sum/units_sum
wash_size = wash_sum/units_sum
tooth_size = tooth_sum/units_sum
soap_size = soap_sum/units_sum
shampoo_size = shampoo_sum/units_sum
lotion_size = lotion_sum/units_sum
```

```python
size = [cream_sum,wash_sum,tooth_sum,soap_sum,shampoo_sum,lotion_sum]
plt.pie(size,labels = features,autopct='%1.1f%%', shadow=True, startangle=140)
```
